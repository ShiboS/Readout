```python

inst.write(":CALC1:PAR1:DEF S21")        # Set measurement of S-parameter
inst.write(":CALC1:FORM POL")            # Set measurement format. "MLOG" log magtitude (dB)
                                         #                         "POL"  polar format  (Re/Im)
                                         #                         "REAL"  real format
                                         #                         "IMAG"  imaginary format
inst.write(":SOUR1:POW -40")             # Set output power (dBm)
inst.write(":SENS1:SWE:POIN 1001")       # Set number of points
inst.write(":SENS1:BAND 10E3")           # Set IF bandwidth (Hz)

inst.write(":SENS1:FREQ:STAR 4.5E9")       # Set start frequency (Hz)
inst.write(":SENS1:FREQ:STOP 4.6E9")       # Set stop frequency (Hz)

inst.write(":SENS1:AVER:COUN 16")        # Set averaging factor
inst.write(":SENS1:AVER ON")             # Enable averaging or disable "OFF"

inst.write(":DISP:WIND1:TRAC1:Y:AUTO")   # Auto scale
inst.write(":OUTP ON")                   # RF out ON or OFF

inst.write(":CALC1:TRAC1:MARK1:ACT")     # Activate Marker1 on Trace 1
```

```python
# we can choose three formats for reading data: 
# "REAL" and "REAL32" are 64bit and 32bit binary seperately
# "d"    and "f"      should be used in "query_binary_values" 
# #ASCii" is ascii format which can be directly read as string
# The final "True" means "Big Endian" which is true for E5071C
#
# With binary format you can directly get float format, which can be treated as number
# With ascii format you can get string format, which should be transformated to float format

# Binary format frequency data query
inst.write(":FORM:DATA REAL")
inst.query_binary_values(":SENS1:FREQ:DATA?", "d", True)
```

```python
# ASCii format frequency data query
inst.write(":FORM:DATA ASCii")
inst.query(":SENS1:FREQ:DATA?")
```

```python
# The trace data contains two numbers for one frequency point.
# When we measure with MAGNITUDE, REAL and IMAGINARY formats, the first number is the value and the second number is ZERO.
# When we measure with POLAR and SMITH formats, the second number contains a value.
# Therefore, we need to save both two values.

# Binary format trace data query
inst.write(":FORM:DATA ASCii")
inst.query(":CALC1:DATA:FDAT?")
```

```python
# ASCii format trace data query
inst.write(":FORM:DATA REAL")
inst.query_binary_values(":CALC1:DATA:FDAT?", "d", True)
```

```python
# Set data transfer format to "REAL" which is 64bit binary
inst.write(":FORM:DATA REAL")
# Readout frequency data
x_data = inst.query_binary_values(":SENS1:FREQ:DATA?", "d", True)
x_array = np.array(x_data)
# Readout S-para data and delete every next data which is "0.0"
y_data = inst.query_binary_values(":CALC1:DATA:FDAT?", "d", True)
```

```python
# Divide trace data to two parts: real and imaginary.
# For magnitude/real/imaginary/phase/SWR format measurement, the imaginary part will be ZERO.
# For polar/smith format measurement, the imaginary part will be imaginary/phase.
y_data_array = np.array(y_data)
y_array_real = y_data_array[0::2]
y_array_imag = y_data_array[1::2]

# Plot both two parts
plt.plot(x_array, y_array_real)
plt.plot(x_array, y_array_imag)
plt.show()
```

```python
import csv
import time

# yyyymmdd string date format
time_yyyymmdd = time.strftime('%Y%m%d')

# Ask VNA what the output power is and we get string format of float value.
# Change this string format to float format, then change it to integar format and then change it back to string format
# u'-4.00000000000E+001\n' ---> -40.0 ---> -40 ---> '-40'
POWER = str(int(float(inst.query(":SOUR1:POW?")))) + 'dBm'

# Start and stop frequencies are the same with power
FSTART = str(int(float(inst.query(":SENS1:FREQ:STAR?")))/1E9)
FSTOP = str(int(float(inst.query(":SENS1:FREQ:STOP?")))/1E9)

# Name of file
name_of_file = time_yyyymmdd + '_' + POWER + '_' +  FSTART + '-' + FSTOP

# Make a new csv format document with name_of_file and write it
f = open(name_of_file + '.csv', 'w')
fwrite = csv.writer(f)

# Write first and second rows with variable names and measurement parameters
fwrite.writerow( ('Frequency', 'S21'))
fwrite.writerow( ('Hz', 'REAL', 'IMAG'))

# Write number-of-points rows
for i in range(0, int(inst.query(":SENS1:SWE:POIN?"))):
    fwrite.writerow([x_array[i], y_array_real[i], y_array_imag[i]])

# Check the file content and print it
contents = open(name_of_file+'.csv', 'r')
a = contents.read()

print a
```
